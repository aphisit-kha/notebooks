```python
# รายการคำนวณดอกเบี้ย
cal_interest_history = []
# วันที่ Accrued ดอกเบี้ยล่าสุด
last_interest_accrued_date = None
last_interest_accrued = {
    'interest_accrued': Decimal('0'),
    'effective_date': None,
}
# last_interest_accrued = {
#     'interest_accrued': Decimal('526.03'),
#     'effective_date': datetime.date(2023, 7, 11),
# }
if last_interest_accrued['effective_date'] is not None:
    actual_date = last_interest_accrued['effective_date'] + timedelta(days=1)
    last_interest_accrued_date = last_interest_accrued['effective_date']
    last_int_balance = {
        'days': 192,
        'interest_round': Decimal('-0.0026029')
    }
    # ดอกเบี้ยค้างรับทั้งหมด ยังไม่รวมการปรับปรุงดอกเบี้ย manual
    interest_days += last_int_balance['days']
    begin_interest_actual += last_interest_accrued['interest_accrued'] + last_int_balance['interest_round']
    begin_interest_accrued += last_interest_accrued['interest_accrued']
    # Accrued ดอกเบี้ยสุทธิ
    # bf_interest_balance = EffInterestBalance.objects.filter(contract_detail_id=contract_detail_id, principal_balance=principal_balance).order_by('effective_date')
    bf_interest_balance = [{
        'start_date': datetime.date(2023, 1, 1),
        'effective_date': datetime.date(2023, 7, 11),
        'principal_balance': Decimal('10000.00'),
        'interest_code': '0',
        'bank_id': None,
        'fix_rate': Decimal('10.0000'),
        'float_rate': Decimal('0.0000'),
        'interest_rate': Decimal('10.0000'),
        'days': 192,
        'interest_accrued': Decimal('526.03'),
        'interest_round': Decimal('-0.0026029'),
        'manual_interest_adjust': Decimal('0.00'),
        'net_interest_accrued': Decimal('526.03'),
    }]
    for foo in bf_interest_balance:
        cal_interest_history.append(dict(
            start_date=foo['start_date'],
            effective_date=foo['effective_date'],
            principal_balance=foo['principal_balance'],
            interest_code=foo['interest_code'],
            bank_id=foo['bank_id'],
            fix_rate=foo['interest_rate'],
            float_rate=foo['float_rate'],
            interest_rate=foo['interest_rate'],
            days=foo['days'],
            interest_actual=foo['interest_accrued']+foo['interest_round'],
            interest_accrued=foo['interest_accrued'],
            interest_round=foo['interest_round'],
            manual_interest_adjust=foo['manual_interest_adjust'],
            net_interest_accrued=foo['net_interest_accrued'],
        ))
    # เริ่มคำนวณดอกเบี้ย
    start_date = actual_date
    # วันที่เริ่มคิดดอกเบี้ย == วันที่ตัดลูกหนี้
    if actual_date == ar_date:
        # คิดดอกเบี้ยถึงวันที่ เท่ากับ วันที่ Accrued ดอกเบี้ยล่าสุด
        interest_actual += begin_interest_actual
        interest_accrued += begin_interest_accrued
    # วันที่ตัดลูกหนี้ back date ต้องปรับปรุงดอกเบี้ยออก
    elif ar_date < actual_date:
        # คิดดอกเบี้ยถึงวันที่ น้อยกว่า วันที่ Accrued ดอกเบี้ยล่าสุด
        interest_actual += begin_interest_actual
        # วันที่เริ่มคิดดอกเบี้ย
        start_date = ar_date
        # อัตราดอกเบี้ยที่มีผลในงวดนี้
        for interest in interest_list:
            if start_date <= last_interest_accrued_date:
                # อัตราดอกเบี้ยที่มีผลช่วงวันที่เริ่มคิดดอกเบี้ย
                if interest['effect_date'] <= start_date <= interest['end_date']:
                    cal_interest = dict()
                    cal_interest['start_date'] = start_date
                    if interest['end_date'] >= last_interest_accrued_date:
                        date_diff = last_interest_accrued_date - start_date
                        # วันที่เริ่มคิดดอกเบี้ย ถัดไป
                        start_date = last_interest_accrued_date + timedelta(days=1)
                    else:
                        date_diff = interest['end_date'] - start_date
                        # วันที่เริ่มคิดดอกเบี้ย ถัดไป
                        start_date = interest['end_date'] + timedelta(days=1)
                    # จำนวนวันที่คิดดอกเบี้ย
                    current_days = date_diff.days + 1
                    current_actual = round((principal_balance * interest['interest_rate'] * current_days) / 36500, 7)
                    cal_interest['interest_actual'] = current_actual
                    # ดอกเบี้ยแท้จริง ยกไป
                    adj_interest_actual += current_actual
                    # ดอกเบี้ยที่คำนวณได้ บวกกลับทศนิยมที่ปัดไว้
                    current_actual += adj_current_round
                    current_accrued = round(current_actual, 2)
                    adj_current_round = current_actual - current_accrued
                    interest_days -= current_days
                    # ดอกเบี้ย Accrued ยกไป
                    adj_interest_accrued += current_accrued
                    # ปรับปรุงดอกเบี้ยออกจากดอกเบี้ยสุทธิที่รับรู้ล่าสุด
                    for bar in cal_interest_history:
                        if bar['interest_code'] == interest['interest_code'] and bar['interest_rate'] == interest['interest_rate']:
                            bar['days'] -= current_days
                            bar['interest_actual'] -= cal_interest['interest_actual']
                            bar['interest_accrued'] -= current_accrued
                            bar['interest_round'] = bar['interest_actual'] - bar['interest_accrued']
                            bar['net_interest_accrued'] -= current_accrued

        interest_actual -= adj_interest_actual
        interest_accrued = round(interest_actual, 2)
        if adj_interest_actual > 0:
            adj_interest_actual = adj_interest_actual * -1
            adj_interest_accrued = adj_interest_accrued * -1
else:
    # เริ่มคำนวณดอกเบี้ย
    start_date = actual_date
```

ตารางสัญญา
